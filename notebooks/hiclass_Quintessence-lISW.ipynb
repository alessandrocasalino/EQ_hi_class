```python
%matplotlib inline

from math import *
import matplotlib
import numpy as np
import matplotlib.transforms as transforms
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import pylab
from scipy import interpolate
import os

a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_lcdm_z2_cl.dat").T
ell_lcdm = a[0]
lens1_lcdm = ell_lcdm*(1.+ell_lcdm)* a[1]
a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_lcdm_cl.dat").T
lens2_lcdm = ell_lcdm*(1.+ell_lcdm)* a[1]

a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_sigma05_alpha01_z2_cl.dat").T
ell = a[0]
lens1_f1 = ell*(1.+ell) * a[1]
a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_sigma05_alpha01_cl.dat").T
lens2_f1 = ell*(1.+ell) * a[1]

a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_sigma05_alphaN01_z2_cl.dat").T
lens1_f2 = ell*(1.+ell) * a[1]
a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_sigma05_alphaN01_cl.dat").T
lens2_f2 = ell*(1.+ell) * a[1]

a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_sigma01_alpha01_z2_cl.dat").T
lens1_f3 = ell*(1.+ell) * a[1]
a = np.genfromtxt("/Users/alessandrocasalino/Software/CLion/EQ_hi_class/output/quin_lISW_sigma01_alpha01_cl.dat").T
lens2_f3 = ell*(1.+ell) * a[1]
```

```python
text_size=42
fig_size_x=32
fig_size_y=24

T0=2.7255

plt.rc('text', usetex=True)
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : text_size}

matplotlib.rc('font', **font)


f,axarr = plt.subplots(2,1,sharey=False, sharex=False,figsize=(fig_size_x,fig_size_y), facecolor='w',gridspec_kw = {'height_ratios':[1, 0.4]})

axarr[0].loglog(ell,lens2_f1,label='$\\sigma=0.5 \,\\alpha=0.1$')
axarr[0].loglog(ell,lens2_f2,label='$\\sigma=0.5 \,\\alpha=-0.1$')
axarr[0].loglog(ell,lens2_f3,label='$\\sigma=0.1 \,\,\,\\alpha=0.1$')
axarr[0].loglog(ell_lcdm,lens2_lcdm,'--',label='$\Lambda$CDM')
axarr[0].title.set_text('$z=3.3$')
axarr[0].set_ylim(1e-11,1e-8)
axarr[0].set_xlim(2,600)
axarr[0].legend()
#axarr[0].title("$k=0.1$")
axarr[0].set_ylabel("$\ell (\ell+1) C_\ell$ $[\mu$K$^2]$")
axarr[0].set_xlabel("$\ell$")

axarr[0].grid(True)

axarr[1].semilogx(ell[:len(ell_lcdm)],(lens2_f1[:len(ell_lcdm)]-lens2_lcdm)/lens2_lcdm)

axarr[1].semilogx(ell[:len(ell_lcdm)],(lens2_f2[:len(ell_lcdm)]-lens2_lcdm)/lens2_lcdm)

axarr[1].semilogx(ell[:len(ell_lcdm)],(lens2_f3[:len(ell_lcdm)]-lens2_lcdm)/lens2_lcdm)

axarr[1].grid(True)

axarr[1].set_ylabel("$\Delta_\ell/C_\ell$")
axarr[1].set_xlabel("$\ell$")

axarr[1].set_xlim(2,600)
axarr[1].set_ylim(-2.5,2.5)

#customticks2=[1e-3,1e-2,1e-1]
#axarr[0].set_xticks(customticks2)

f.subplots_adjust(wspace=0)
f.subplots_adjust(hspace=0)
```
