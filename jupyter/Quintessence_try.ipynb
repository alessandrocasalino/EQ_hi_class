```python
%matplotlib inline

from math import *
import matplotlib
import numpy as np
import matplotlib.transforms as transforms
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

from matplotlib import *
from classy import Class

plt.rc('text', usetex=True)
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 26}

matplotlib.rc('font', **font)
```

```python
filename = '/Users/Alessandro/Desktop/Programmi/hi_class/output/ini/quintessence_class00_background.dat'
data1=np.genfromtxt(filename, delimiter='', comments='#');

l=data1[:,0]
d1=data1[:,10]
crit=data1[:,13]
d2=data1[:,12]
d3=data1[:,14]

f1 = plt.figure(figsize=(24,10))

colors=('red','blue','green','yellow','magenta','brown','black');

ax1 = f1.add_subplot(1,1,1)
ax1.loglog(l,d1/crit,color=colors[0])
ax1_legend = mpatches.Patch(color=colors[0], label='$\Omega_{CDM}$')

ax2= f1.add_subplot(1,1,1)
ax2.loglog(l,d2/crit,color=colors[1])
ax2_legend = mpatches.Patch(color=colors[1], label='$\Omega_{r}$')

ax3 = f1.add_subplot(1,1,1)
ax3.loglog(l,d3/crit,color=colors[2])
ax3_legend = mpatches.Patch(color=colors[2], label='$\Omega_{\phi}$')

plt.axis([0,pow(10,7),pow(10,-22),10])

plt.legend(handles=[ax1_legend,ax2_legend,ax3_legend])
#plt.title("Fractional density of some components")
plt.xlabel("z")
plt.ylabel("$\Omega$")

plt.show()
```
